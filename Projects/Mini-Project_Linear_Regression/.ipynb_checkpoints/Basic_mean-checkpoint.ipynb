```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
mean = diamonds['price'].mean()
diamonds['price_predicted']= mean
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(diamonds.price,diamonds.price_predicted)
```

```python
mean_squared_error(diamonds.price,diamonds.price_predicted)
```
